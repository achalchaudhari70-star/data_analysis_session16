```python
import seaborn as sb
plt.subplot(2,2,1)
sb.histplot(df["age"])
plt.title("Distribution of AGE")
plt.subplot(2,2,2)
sb.histplot(df["bmi"])
plt.title("Distribution of BMI")
plt.subplot(2,2,3)
sb.histplot(df["children"])
plt.title("Distribution of CHILDREN ")
plt.subplot(2,2,4)
sb.histplot(df["expenses"])
plt.title("Distribution of EXPENSES")
plt.tight_layout()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sb.countplot(x="sex",data=df)
plt.title("count plot of SEX")
plt.xlabel("sex")
plt.ylabel("count")
plt.subplot(1,3,2)
sb.countplot(x="smoker",data=df)
plt.title("count plot of SMOKER")
plt.xlabel("SMOKER")
plt.ylabel("count")
plt.subplot(1,3,3)
sb.countplot(x="region",data=df)
plt.title("count plot of REGION")
plt.xlabel("region")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```

```python
corr=df.corr(numeric_only=True)
sb.heatmap(corr,annot=True ,cmap="mako")
plt.title("correlation heatmap of numeric value")
```

```python
avg_expenses=df["expenses"].mean()
print("Avarage insurence expence:",avg_expenses)
max_expenses=df["expenses"].max()
print("Maximum insurence expenses:",max_expenses)
min_expenses=df["expenses"].min()
print("Minimum insurence expenses:",min_expenses)
avg_expenses_by_smoker=df.groupby("smoker")["expenses"].mean()
print("Avarage expenses by smoker status:",avg_expenses_by_smoker)
```

```python
plt.figure(figsize=(6,5))
sb.boxplot(x="smoker",y="expenses",data=df,hue="sex",palette="rocket")
plt.title("Distribution of insurence expenses by smoker")
plt.xlabel("smoker")
plt.ylabel("insurence expenses")
```
